```python
x=range(1,24)
y=[random.randint(0,40) for i in x]
plt.figure(figsize=(20,10),dpi=100)

# 设置X轴
x_ticks_labels=[f"{i}:00" for i in x]
plt.xticks(x,x_ticks_labels)

# 设置Y轴
y_range = range(min(y), max(y)+1)
y_ticks_labels=[f"{j}C°" for j in y_range]
plt.yticks(y_range, y_ticks_labels)

plt.plot(x,y)
plt.show()
```

```python
from matplotlib import font_manager
x=range(1,24)
y=[random.randint(0,40) for i in x]
plt.figure(figsize=(20,10),dpi=100)
my_font = font_manager.FontProperties(fname="C:/Windows/Fonts/STKAITI.TTF",size=20)

# 设置X轴
x_ticks_labels=[f"{i}点" for i in x]
plt.xticks(x,x_ticks_labels,fontproperties=my_font)

# 设置Y轴
y_range = range(min(y), max(y)+1)
y_ticks_labels=[f"{j}摄氏度" for j in y_range]
plt.yticks(y_range, y_ticks_labels,fontproperties=my_font)

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x = np.arange(1, 100)

# 创建 2x2 的子图布局
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 获取各个子图坐标轴对象
ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]

# 分别绘图
ax1.plot(x, x)          # 线性
ax2.plot(x, -x)         # 反向线性
ax3.plot(x, x**2)       # 平方（抛物线）
# 为第三个子图添加红色虚线网格
ax3.grid(color='r', linestyle='--', linewidth=1, alpha=0.3)

ax4.plot(x, np.log(x))  # 对数

plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random

x=range(1,31)
y1=[random.randint(20,35) for i in x]

plt.figure(figsize=(20,10),dpi=100)

# 修正：直接使用 plt.scatter，并使用正确的变量 y1
plt.scatter(x,y1,label="三月份")

my_font=font_manager.FontProperties(fname="C:/Windows/Fonts/STKAITI.TTF",size=20)

x_ticks_labels=["3月{}号".format(i) for i in x]

plt.xticks(x,x_ticks_labels,fontproperties=my_font, rotation=45)

# 修正：去掉等号，直接调用函数
plt.xlabel("日期",fontproperties=my_font)
plt.ylabel("温度",fontproperties=my_font)

plt.legend(prop=my_font)

plt.show()
```
