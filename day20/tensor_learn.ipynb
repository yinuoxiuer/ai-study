```python
"""
GPU中tensor的使用:
1.实例化device. torch.device("cpu" or "cuda:0"),选cuda:0需要torch.cuda.is_available()==true
2.tensor.to(device),把tensor转化为CUDA支持的tensor或者CPU支持的tensor
"""
print(torch.cuda.is_available())
if torch.cuda.is_available():
    device = torch.device("cuda") #cuda device对象
    y = torch.ones_like(x, device = device) #创建一个在cuda上的tensor
    x = x.to(device) #把x转为cuda上的tensor
    z = x+y
    print(z)
    print(z.to("cpu", torch.double))

# torch.tensor([1.9806], device = "cuda:0")
device = torch.device("cpu")
x.to(device)
```

张量乘法是hadmard乘积

```python
import torch
import numpy as np

# 1. 创建两个形状相同的张量 (3行4列)
# np.arange(12) 生成 0到11 的数字
# reshape(3, 4) 变成 3x4 矩阵
t8 = torch.tensor(np.arange(12).reshape(3,4), dtype=torch.float32)
t9 = torch.tensor(np.arange(12).reshape(3,4), dtype=torch.float32)

# 2. 元素对应相乘 (Element-wise Multiplication)
# 对应位置的数相乘，例如 [0,0]位置 * [0,0]位置
print(t8 * t9)
# 结果形状依然是 (3, 4)

# 3. 矩阵乘法 (Matrix Multiplication)
# torch.mm 是 Matrix Multiply 的缩写
# t8 的形状是 (3, 4)
# t9 的形状是 (3, 4)，这两个不能直接做矩阵乘法！
# 必须把 t9 转置 (transpose) 成 (4, 3)
# (3, 4) 乘以 (4, 3) -> 结果形状是 (3, 3)
print(torch.mm(t8, t9.transpose(0, 1)))
```
