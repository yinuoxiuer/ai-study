```python
import numpy as np
import random
a=np.random.randint(0, 100, size=(6, 6))
a = a.astype(float)
a[3,4]=np.nan
a[5,5]=np.nan
print(a)
for i in range(a.shape[1]):
    temp_col=a[:,i]
    nan_num=np.count_nonzero((temp_col!=np.nan))
    if nan_num!=0:
          temp_col_not_nan = temp_col[temp_col == temp_col]
          temp_col[np.isnan(temp_col)] = np.mean(temp_col_not_nan)
print(a)

```
